```python
import networkx as nx
```

```python
G = nx.DiGraph()
```

```python
G.add_node('Alice')
G.add_node('Bob')
G.add_node('Cruck')
```

```python
list(G.nodes())
```

```python
G.add_edge('Alice', 'Bob')
G.add_edge('Bob', 'Alice')

G.add_edge('Bob', 'Cruck')
G.add_edge('Alice', 'Cruck')
```

```python
list(G.edges())
```

```python
COLORS = ['red', 'green', 'blue']
nx.draw_circular(G,
                 node_color=COLORS[0],
                 node_size=2000,
                 with_labels=True)
```

```python
positions = dict(Albany=(-74, 43),
                 Boston=(-71, 42),
                 NYC=(-74, 41),
                 Philly=(-75, 40))
```

```python
drive_times = {('Albany', 'Boston'): 3,
               ('Albany', 'NYC'): 4,
               ('Boston', 'NYC'): 4,
               ('NYC', 'Philly'): 2}
```

```python
G = nx.Graph()
G.add_nodes_from(positions)
G.add_edges_from(drive_times)
```

```python
nx.draw(G,
        positions,
        node_color=COLORS[0],
        node_shape='s',
        node_size=2500,
        with_labels=True)
```

```python
nx.draw_networkx_edge_labels(G,
                         positions,
                         edge_labels=drive_times)
```

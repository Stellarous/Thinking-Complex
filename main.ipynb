```python
import networkx as nx
```

```python
G = nx.DiGraph()
```

```python
G.add_node('Alice')
G.add_node('Bob')
G.add_node('Cruck')
```

```python
list(G.nodes())
```

```python
G.add_edge('Alice', 'Bob')
G.add_edge('Bob', 'Alice')

G.add_edge('Bob', 'Cruck')
G.add_edge('Alice', 'Cruck')
```

```python
list(G.edges())
```

```python
COLORS = ['red', 'green', 'blue']
nx.draw_circular(G,
                 node_color=COLORS[0],
                 node_size=2000,
                 with_labels=True)
```
